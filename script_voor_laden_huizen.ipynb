### Scriptje voor gemakkelijk inladen van huizen data

```python
import os
import pandas as pd
```

```python
root = os.getcwd()
huizen = os.listdir(os.path.join(root, 'huizen'))
huizen = [os.path.join(root, 'huizen', huis) for huis in huizen]
```

```python
# alle data wordt geladen in een pandas dataframe
df:pd.DataFrame = None

for huis in huizen:
    
    huis_df:pd.DataFrame = pd.read_csv(huis, sep=';')

    if df is None:
        df = huis_df
    else:
        df = pd.concat([df, huis_df], axis=0)
```
